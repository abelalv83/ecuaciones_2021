```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
from IPython.html.widgets import interact
```

```python
def fun(y,t,k):
    dydt = y*(1-y)+k
    return dydt

# initial condition


# time points
def campo(xi,xf,N,c):
  x = np.linspace(xi, xf, N)
  y = np.linspace(-2, 2, N)
  X, Y = np.meshgrid(x, y)
  U = 1
  V = fun(Y,X,c) 
  U2, V2 = U/np.sqrt(U**2+V**2), V/np.sqrt(U**2+V**2)
  plt.quiver(X, Y, U2, V2, color='k')
  plt.grid(True)
  plt.title(r'Campo de pendientes: $\frac{dy}{dx}=y(1-y)+$c')
  return


def gra(c):
  N=30

  xi=-1
  xf=1
  # solve ODEs
  #k = 0.1
  t = np.linspace(xi,xf)
  y0=[2.0,0.5,0.9,0.2]
  for yi in y0:
    y1 = odeint(fun,yi,t,args=(c,))
    plt.plot(t,y1,linewidth=2)
  # plot results
  campo(xi,xf,N,c)
  
  return
interact(gra,c=(-0.3,2,0.1))




```
