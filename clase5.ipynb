# Ejemplo 1
solución de la ecuación diferencial
$$3xy'(x)-(x^2-9)y(x)+\frac{1}{x}=0$$

Para resolver esto lo primero que ahcemos es imnportar la librerias sympy
mediante el código 

>import sympy as sy

podemos usar este comando para mejorar la impresión de las repuestas

>sy.init_printing(use_latex='mathjax') 

Ahora definimos la variable independiente, para nuestro caso es x

>x = sy.Symbol('x')

y la varaible depéndiente 

> y = sy.Function('y')

podemos llamar la derivada como 

> dy=y(x).diff(x)

Escribimos la ecuación, $3xy'(x)-(x^2-9)y(x)+\frac{1}{x}=0$
(note que en lugar de emplear el igual usamos la coma)

>sy.Eq(3*x*dy-(x**2-9)*y(x)+1/x,0)



```python
# Otra manera y desplegando mas informacion de la EDOpython
import sympy as sy
sy.init_printing(use_latex='mathjax') # para mejor despliege de la EDO


# aqui definimos la variables Note que y es un afunción y x es la variable independiente
x = sy.Symbol('x')
y = sy.Function('y')
# Definicion de E.D a resolver:


# llamaremos la derivada 
dy=y(x).diff(x)

sy.Eq(3*x*dy-(x**2-9)*y(x)+1/x,0)



```

para resolver el problema usamos las dos líneas de código que se encuentran en la parte inferior

```python
q=sy.Eq(3*x*dy-(x**2-9)*y(x)+1/x,0)
sy.dsolve(q,y(x))
```

# Problemas de valor inicial  con Sympy
$y'+3y=6t+5$ con $y(0)=3$


```python
# aqui definimos la variables Note que y es un afunción y x es la variable independiente
t = sy.Symbol('t')
y = sy.Function('y')
q=sy.Eq(y(t).diff(t)+3*y(t),6*t+5)
```

Para resolver las condiciones inicales agregamos al dsolve el comando ics={}
que el comando para las condiciones iniciales 

```python
CI={y(0):3}
sy.dsolve(q,y(t),ics=CI)
```

```python
# podemos simplificar la expresión

sy.simplify(sy.dsolve(q,y(t),ics=CI))
```
