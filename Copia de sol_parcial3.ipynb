```python
import numpy as np
import sympy as sy
sy.init_printing()  # imprimir formato latex
def calculo(A):
  m=sy.Matrix(A)
  display(m)
  m=sy.Matrix(A)
  print('Los autovalores')
  display(m.eigenvals())
  print('Los autovectores')
  display(m.eigenvects())
  print('******verificación*******')
  L, V = np.linalg.eig(A)
  print('autivalores',L,'autovectores',V)
  
  P, J = m.jordan_form()

  print('J=')
  display(J)
  print('P=')
  display(P)
  
  return
 
```

```python
# codigo para sistema 2x2
def sol2x2(A):
	AA=sy.Matrix(A)
	t,C1,C2 = sy.symbols("t C1 C2")
	x,y = sy.symbols("x y", cls=sy.Function)
	N=2

	X=sy.Matrix([x(t), y(t)])
	F=AA*X
	Eq1=sy.Eq(X.diff(t),F)
	print("Ecuacion")
	display(Eq1)

	Eq1 = sy.Eq(sy.diff(x(t), t), F[0])
	Eq2 = sy.Eq(sy.diff(y(t), t), F[1])

	eqs = [Eq1, Eq2]

	#dsolve_system(eqs)
	# Solutions with undetermined constants of integration, C1, C2 and C3:
	soln = sy.dsolve(eqs)
	display(soln)
	
	# Calculation of the above constants:
	soln = sy.dsolve(eqs, ics = {x(0): 1, y(0): 1})
	for j in range(N):
	  display(soln[j].rhs)
	  sy.plot(soln[j].rhs,(t,-1,10))

  
```

```python
# codigo para sistema 3x3
def sol3x3(A):
	t,C1,C2,C3 = sy.symbols("t C1 C2 C3")
	x,y,z = sy.symbols("x y z", cls=sy.Function)
	N=3

	X=sy.Matrix([x(t), y(t),z(t)])
	A=sy.Matrix(A)
	print(A)

	F=A*X
	Eq1=sy.Eq(X.diff(t),F)

	display(Eq1)

	Eq1 = sy.Eq(sy.diff(x(t), t), F[0])
	Eq2 = sy.Eq(sy.diff(y(t), t), F[1])
	Eq3 = sy.Eq(sy.diff(z(t), t), F[2])
	eqs = [Eq1, Eq2, Eq3]

	#dsolve_system(eqs)
	# Solutions with undetermined constants of integration, C1, C2 and C3:
	soln = sy.dsolve(eqs)
	display(soln)
	# Calculation of the above constants:
	soln = sy.dsolve(eqs, ics = {x(0): 1, y(0): 1, z(0): 1})
	for j in range(N):
	  display(soln[j].rhs)
	  sy.plot(soln[j].rhs,(t,-1,50))
```
