## Ejercicio de tanques

Consideremos un gran tanque que contiene azúcar y agua con lo que se preparán refrescos embotellados. Supogase que 

  a. El tanque contiene 100 galones de líquido. Además, la cantidad que fluye hacia adentro es la misma que fluye hacia afuera, pero siempre hay 100 galones en el tanque. 
  
  b. El tanque se mantiene bien mezclado, por lo que la concentración de azucar es uniforme en todo el tanque.

  c. El agua azucarada que contiene 5 cucharadas de azúcar por galon entra al tanque a trvéz de un tubo A a razón de 2 galones por minuto 

  d. El agua azucarada que contiene 10 cucharadas de azúcar por galón  entra al tanque a travéz del tubo B a razón 1 galón por minuto

  e. El agua azucarada sale del Tanque a través del tubo C a razón de 3 galones por minuto

```python
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import numpy as np


# Ecuacion Diferencial: dy/dx=f(x,y)
        
# Ecuacion Diferencial: dy/dx=f(x,y)
        


def f( y,x): 
    
    return ((y**7)*((1-y/20.0)**3)*(y/5.0-1))
    



# Campo de pendientes
N = 30  # numero de pendientes a graficar
xi = 0
xf = 2.0  # dominio
x = np.linspace(xi, xf, N)
y = np.linspace(-5, 25, N)
X, Y = np.meshgrid(x, y)
U = 1
V = f(Y,X)
U2, V2 = U/np.sqrt(U**2+V**2), V/np.sqrt(U**2+V**2)
plt.quiver(X, Y, U2, V2, color='k')
plt.grid(True)
plt.title(r'Campo de pendientes: $\frac{dy}{dx}=y(1-y)$')




```

```python
# initial condition
y0 = [0.0, -3,3, 15,25]

# time points
t = np.linspace(0, xf)

# solve ODE
for yi in y0:
    y = odeint(f, yi, t, rtol=1.0E-8)
    plt.plot(t, y)

# plot results

plt.xlabel('time')
plt.ylabel('y(t)')
plt.quiver(X, Y, U2, V2, color='k')
plt.grid(True)
plt.title(
    r'Campo de pendientesny solución de la ecuación: $(1-\frac{y}{20})^3(\frac{y}{5}-1)y^7$')
```
