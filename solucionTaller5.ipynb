1. Resuelva las siguientes ecuaciones diferenciales 

  a. $y'=5y(1-y)$
  
  b. $y'=ky(100-y)$, donde $k$ es una constante. Sujeto a $y(0)=5$ y $y(2)=15$

```python
x = sy.Symbol('x')
y = sy.Function('y')
dy=y(x).diff(x)

sy.Eq(dy,5*y(x)*(1-y(x)))
q=sy.Eq(dy,5*y(x)*(1-y(x)))
sy.dsolve(q,y(x))


```

```python
k = sy.Symbol('k')
sy.Eq(dy,k*y(x)*(100-y(x)))
q=sy.Eq(dy,k*y(x)*(100-y(x)))
expre=sy.dsolve(q,y(x))
print("solucion general",expre)
expre2=expre.subs(x,0)
C=sy.Symbol('C')
expre3=sy.solve(1/(C+1)-5,C)
print("c=",expre3)
expre.subs(x,2)



```

solución de la ecuación diferencial
$$3xy'(x)-(x^2-9)y(x)+\frac{1}{x}=0$$

```python
# Otra manera y desplegando mas informacion de la EDOpython
import sympy as sy
sy.init_printing(use_latex='mathjax') # para mejor despliege de la EDO


# aqui definimos la variables Note que y es un afunción y x es la variable independiente
x = sy.Symbol('x')
y = sy.Function('y')
# Definicion de E.D a resolver:


# llamaremos la derivada 
dy=y(x).diff(x)

sy.Eq(dy,y)



```

# Problemas de valor inicial  con Sympy
$y'+3y=6t+5$ con $y(0)=3$


```python
# Otra manera y desplegando mas informacion de la EDOpython
import sympy as sy
sy.init_printing(use_latex='mathjax') # para mejor despliege de la EDO
# aqui definimos la variables Note que y es un afunción y x es la variable independiente
t = sy.Symbol('t')
y = sy.Function('y')
q=sy.Eq(y(t).diff(t),20-0.03*y(t))
```

Para resolver las condiciones inicales agregamos al dsolve el comando ics={}
que el comando para las condiciones iniciales 

```python
CI={y(0):3}
sy.dsolve(q,y(t),ics=CI)
```

```python
# podemos simplificar la expresión

sy.simplify(sy.dsolve(q,y(t),ics=CI))
```
