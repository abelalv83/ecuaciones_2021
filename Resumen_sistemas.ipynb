# Solución de sistemas de ecuaciones diferenciales lineales 

Sea el siguiente sistema de ecuaciones diferenciales lineales 

$$x'(t)=Ax(t)$$

donde $\left[\begin{matrix}
x_1(t)\\x_2(t)
\end{matrix}\right]
$ y la matriz 

$$A= \left[\begin{matrix}
a_{11}&a_{12}\\a_{21}&a_{22}
\end{matrix}\right]$$
donde los elementos de $a_{ij}$ son constantes. 

Para encontrar la solución del sistema debemos estudiar sus autovalores y autovectores, pues segun multiplicidad podemos cálcular la solución del problema. 

## Cuando la multiplicidad algebraica y geometrica coinciden 

Este caso sucede cuando tienen 

* Dos autovalores reales diferentes
* Tiene un autovalor real pero dos autovectores 
* Tienen autovalores complejos




### **Caso 1** Dos autovalores reales diferentes 

Cuando una matriz $2\times 2$ tiene dos autovalores reales diferentes esta se puede diagonalizar, es decir supongamos que $A$ tiene  los autovalores 
$\lambda_1$ y $\lambda_2$ con los autovectores asociados $v_1,v_2$, respectivamente 
$$A= \left[\begin{matrix}
a_{11}&a_{12}\\a_{21}&a_{22}
\end{matrix}\right]=V\left[\begin{matrix}
\lambda_1&\\&\lambda_2
\end{matrix}\right]V^{-1},$$
donde $V=[v_1,v_2]$.


La solcuión del sistema de ecuaciones diferenciales 
$$x'=Ax$$ esta determindo por 

$$x(t)=\left[\begin{matrix}
x_1(t)\\x_2(t)
\end{matrix}\right]=C_1 v_1e^{\lambda_1t}+C_2 v_2e^{\lambda_2t}
$$

#### **Ejemplo**


Sea el sistema de ecuaciones 

$$\begin{cases}x'(t)&=5x(t)+3y(t)\\y'(t)&=-6x(t)-4y(t),\end{cases}$$

Note que este sistema se puede escribir de la forma 


$$\frac{d}{dt}\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right]=\left[\begin{matrix}
5&3\\-6&-4
\end{matrix}\right]\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right],$$

donde la matriz $A=\left[\begin{matrix}
5&3\\-6&-4
\end{matrix}\right]$ tiene dos autovalores $\lambda_1=-1$ y $\lambda_2=2$
con sus respectivos autovectores $v_1=\left[\begin{matrix}
1\\-2
\end{matrix}\right]$ y  $v_2=\left[\begin{matrix}
1\\-1
\end{matrix}\right].$ Deesta for,a podemos definir 

$$V=\left[\begin{matrix}
1&1\\-2&-1
\end{matrix}\right]$$ y tenemos que $$V^{-1}=\left[\begin{matrix}
-1&-1\\2&1
\end{matrix}\right].$$

Note que $$A=\left[\begin{matrix}
5&3\\-6&-4
\end{matrix}\right]=\left[\begin{matrix}
1&1\\-2&-1
\end{matrix}\right]\left[\begin{matrix}
-1&\\&2
\end{matrix}\right]\left[\begin{matrix}
-1&-1\\2&1
\end{matrix}\right].$$
 sustituyendo esto en el sistema obtenemos 


 
$$\frac{d}{dt}\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right]=\left[\begin{matrix}
1&1\\-2&-1
\end{matrix}\right]\left[\begin{matrix}
-1&\\&2
\end{matrix}\right]\left[\begin{matrix}
-1&-1\\2&1
\end{matrix}\right]\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right],$$

multiplicando la igualda por $V^{-1}$

$$\frac{d}{dt}\left[\begin{matrix}
-1&-1\\2&1
\end{matrix}\right]\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right]=\left[\begin{matrix}
-1&\\&2
\end{matrix}\right]\left[\begin{matrix}
-1&-1\\2&1
\end{matrix}\right]\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right],$$

haciendo 

$$w(t)=\left[\begin{matrix}
w_1(t)\\w_2(t)
\end{matrix}\right]=\left[\begin{matrix}
-1&-1\\2&1
\end{matrix}\right]\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right]$$

obteneomos el sistema 


$$\frac{d}{dt}\left[\begin{matrix}
w_1(t)\\w_2(t)
\end{matrix}\right]=\left[\begin{matrix}
-1&\\&2
\end{matrix}\right]\left[\begin{matrix}
w_1(t)\\w_2(t)
\end{matrix}\right],$$
el cual se reduce a un sistema desacoplado 

$$\begin{cases}w_1'(t)=-w_1(t)\\w_2'(t)=2w_2(t)\end{cases}$$
de esta forma tenemos la solución 
$$\begin{cases}w_1(t)=C_1e^{-t}\\w_2(t)=C_2e^{2t}\end{cases}$$

sustituyendo en 

$$\left[\begin{matrix}
C_1e^{-t}\\C_2e^{2t}
\end{matrix}\right]=\left[\begin{matrix}
-1&-1\\2&1
\end{matrix}\right]\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right]$$


de esta forma tenemos que la solución es 
$$\left[\begin{matrix}
x(t)\\y(t)\end{matrix}\right]=C_1e^{-t}\left[\begin{matrix}
1\\-2
\end{matrix}\right]+C_2e^{2t}\left[\begin{matrix}
1\\-1
\end{matrix}\right]
$$

## **Casos 2**  un  autovalor real y dos autovalores  

Sea el sistema de 
$$\begin{cases}x'(t)=x(t),\\ 
y'(t)=y(t),\end{cases}$$
este se puede escribir de la forma 
$$\frac{d}{dt}\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right]=\left[\begin{matrix}
1&0\\0&1
\end{matrix}\right]\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right].$$
Note que la  matriz $A=\left[\begin{matrix}
1&0\\0&1
\end{matrix}\right]$ tiene un autovalor y dos atovectores, el autovalor $\lambda=1$ y dos autovectores $v_1=\left[\begin{matrix}
1\\0
\end{matrix}\right]$ y $v_2=\left[\begin{matrix}
0\\1
\end{matrix}\right]$, por tanto la solución del sistema 



es 

$$\begin{split}\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right]&=C_1e^{\lambda t}v_1+C_2e^{\lambda t}v_2,\\
&=C_1e^{t}\left[\begin{matrix}
0\\1
\end{matrix}\right]+C_2e^{t}\left[\begin{matrix}
1\\0
\end{matrix}\right].\end{split}$$

note que aui porcedemos igual que el **caso 1** y esto es porque la matriz $A$ se puede diagonalizar.

## **Casos 3**  Dos autovalores Complejos 

Sea el sistema 

$$\begin{cases}x'(t)=6x(t)-y,\\ 
y'(t)=5x+4y(t),\end{cases}$$

este se puede escribir de la forma 
$$\frac{d}{dt}\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right]=\left[\begin{matrix}
6&-1\\5&4
\end{matrix}\right]\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right].$$
Note que la  matriz $A=\left[\begin{matrix}
6&-1\\5&4
\end{matrix}\right]$
en el caso que la matriz A tiene dos autovalores complejos,
puesto que 

$$det(A-\lambda I)=\left|\begin{matrix}
6-\lambda &-1\\
 5& 4-\lambda
\end{matrix}\right|=\lambda^2-10\lambda +29,$$
las raíces de este polinomio son $\lambda_1=5+2i$ y $\lambda_2=5-2i$, lo cual quiere decir que uno es el conjugado del otro, es decir ellos tienen la forma $\lambda_1=\alpha+i\beta$ y $\lambda_2=\alpha-i\beta$, de aqui notamos que los autovectores tambien son números complejos. Recordemos el precedimiento para encontrar un autovector 
$$\begin{split}Av_1&=\lambda_1 v_1\\
(A-\lambda_1 I)v_1&=0\end{split}$$
Sustituyendo en la matriz $A$ y el valor de $\lambda_1$ en la ecuación anterior

$$\begin{split}
\left[\begin{matrix}
6-\lambda_1 &-1\\
 5& 4-\lambda_1
\end{matrix}\right]\left[\begin{matrix}
v_{11}\\
 v_{12}
\end{matrix}\right]&=\left[\begin{matrix}
0\\
 0
\end{matrix}\right]\\ 
\left[\begin{matrix}
6-(5+2i) &-1\\
 5& 4-(5+2i)
\end{matrix}\right]\left[\begin{matrix}
v_{11}\\
 v_{12}
\end{matrix}\right]&=\left[\begin{matrix}
0\\
 0
\end{matrix}\right]\\
\left[\begin{matrix}
1+2i &-1\\
 5& -(1+2i)
\end{matrix}\right]\left[\begin{matrix}
v_{11}\\
 v_{12}
\end{matrix}\right]&=\left[\begin{matrix}
0\\
 0
\end{matrix}\right]
\end{split}$$
Note que el $det\left(\left[\begin{matrix}
1+2i &-1\\
 5& -(1+2i)
\end{matrix}\right]\right)=0$, por tanto el sistema se reduce a una sola ecuación, así

$$(1-2i)v_{11}-v_{12}=0,$$
de esta forma si tomamos $v_{11}=1$ tenemos que 
$v_{12}=1-2i$, asi el autovector 
$$\begin{split}v_1&=\left[\begin{matrix}
1\\
 1-2i
\end{matrix}\right]\\&=\left[\begin{matrix}
1\\
 1
\end{matrix}\right]-i\left[\begin{matrix}
0\\
 2
\end{matrix}\right]\end{split}.$$
Realizando un procedimietno similar tenemos que el autovector asociado al autovalor $\lambda_2=5-2i$
$$\begin{split}v_2&=\left[\begin{matrix}
1\\
 1+2i
\end{matrix}\right]\\&=\left[\begin{matrix}
1\\
 1
\end{matrix}\right]+i\left[\begin{matrix}
0\\
 2
\end{matrix}\right]\end{split}.$$

Note que $v_1=\overline{v_2}$. Siguiendo el procedimiento del **caso 1** vemos que la matriz $A$ es daognalizable en el campo de los complejos, de esta forma tendriamos la solución 


de esta forma tenemos que la solución es 
$$\begin{split}\left[\begin{matrix}
x(t)\\y(t)\end{matrix}\right]&=C_1v_1e^{\lambda_1 t}+C_2v_2e^{\lambda_2 t}\\
&=C_1e^{(5+2i)t}\left[\begin{matrix}
1\\1-2i
\end{matrix}\right]+C_2e^{(5-2i)t}\left[\begin{matrix}
1\\1+2i
\end{matrix}\right]\\
&=C_1e^{(5+2i)t}\left(\left[\begin{matrix}
1\\
 1
\end{matrix}\right]-i\left[\begin{matrix}
0\\
 2
\end{matrix}\right]\right)+C_2e^{(5-2i)t}\left(\left[\begin{matrix}
1\\
 1
\end{matrix}\right]+i\left[\begin{matrix}
0\\
 2
\end{matrix}\right]\right)\end{split}
$$

como nosotros queremos encontrar la solución real del sistema, usaremos la identidad $e^{(\alpha+i\beta)}=e^{\alpha}(\cos\beta+i\sin\beta)$, luego 
$$\begin{split}\left[\begin{matrix}
x(t)\\y(t)\end{matrix}\right]&=e^{5t}\left(C_1(\cos(2t)+i\sin(2t))\left(\left[\begin{matrix}
1\\
 1
\end{matrix}\right]-i\left[\begin{matrix}
0\\
 2
\end{matrix}\right]\right)+C_2(\cos(2t)-i\sin(2t))\left(\left[\begin{matrix}
1\\
 1
\end{matrix}\right]+i\left[\begin{matrix}
0\\
 2
\end{matrix}\right]\right)\right)\end{split}
$$



**Recordemos** que la multiplicación de dos numeros complejos es de la siguiente forma:
$$(a_1+ib_1)*(a_2+ib_2)=a_1a_2-b_1b_2+i(a_1b_2+a_2b_1)$$

de esta forma tenemos que 

$$\begin{split}C_1(\cos(2t)+i\sin(2t))\left(\left[\begin{matrix}
1\\

\end{matrix}\right]-i\left[\begin{matrix}
0\\
 2
\end{matrix}\right]\right)&=C_1\left\{\left(\cos(2t)\left[\begin{matrix}

\end{matrix}\right]+\sin(2t)\left[\begin{matrix}

\end{matrix}\right] \right)-i\left(\cos(2t)\left[\begin{matrix}
0\\
 2
\end{matrix}\right]-\sin(2t)\left[\begin{matrix}
1\\
 1
\end{matrix}\right] \right)\right\}\end{split}$$


de la misma forma 


$$\begin{split}C_2(\cos(2t)-i\sin(2t))\left(\left[\begin{matrix}
1\\
 1
\end{matrix}\right]-i\left[\begin{matrix}
0\\
 2
\end{matrix}\right]\right)&=C_2\left\{\left(\cos(2t)\left[\begin{matrix}
1\\
 1
\end{matrix}\right]-\sin(2t)\left[\begin{matrix}
0\\
 2
\end{matrix}\right] \right)-i\left(\cos(2t)\left[\begin{matrix}
0\\
 2
\end{matrix}\right]+\sin(2t)\left[\begin{matrix}
1\\
 1
\end{matrix}\right] \right)\right\}\end{split}$$

de esta forma la parte real se tiene 

$$\begin{split}real\left(\left[\begin{matrix}
x(t)\\y(t)\end{matrix}\right]\right)&=e^{5t}\left\{C_1\left(\cos(2t)\left[\begin{matrix}
1\\
 1
\end{matrix}\right]+\sin(2t)\left[\begin{matrix}
0\\
 2
\end{matrix}\right] \right)+C_2\left(\cos(2t)\left[\begin{matrix}
1\\
 1
\end{matrix}\right]-\sin(2t)\left[\begin{matrix}
0\\
 2
\end{matrix}\right] \right)\right\}\end{split}
$$
**Note** que esta forma si tenemos un sistema EDO lieal homogeneo con coeficientes constantes lo podemos escribir la solcuón de la forma 
$B_1=real(v_1)$ y $B_2=img(v_1)$


De la misma forma tenemos la parte imaginaria de la ecuación es 
$$\begin{split}Img\left(\left[\begin{matrix}
x(t)\\y(t)\end{matrix}\right]\right)&=-e^{5t}\left\{C_1\left(\cos(2t)\left[\begin{matrix}
0\\
 2
\end{matrix}\right]-\sin(2t)\left[\begin{matrix}
1\\
 1
\end{matrix}\right] \right)+C_2\left(\cos(2t)\left[\begin{matrix}
0\\
 2
\end{matrix}\right]+\sin(2t)\left[\begin{matrix}
1\\
 1
\end{matrix}\right] \right)\right\}\end{split}
$$






De esta forma podemos tener podemos construir 
$$X_1=\Big[B_1\cos\beta t-B_2\sin\beta t\Big]e^{\alpha t}$$
$$X_2=\Big[B_2\cos\beta t+B_2\sin\beta t\Big]e^{\alpha t}$$

y así construir la solución  general 
$$X=C_1X_1+C_2X_2$$
donde B_1=$real(v_1)$ y $B_2=img(v_1)$ e $\lambda=\alpha+i\beta$

# Caundo la multipliciada algebraica y geometrica no coinciden

Este caso es cuando tiene un único eigenvalor y un único eigenvector

Un ejemplo de ello es el sistema 
$$\begin{cases}
x'= 3x-18y\\
y'=2x-9y
\end{cases}$$

Note que el sistema se puede escribir de la forma 

$$\frac{d}{dt}\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right]=\left[\begin{matrix}
3&-18\\2&-9
\end{matrix}\right]\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right],$$

Asi vemos que el polinomio caracteristico esta determinado por 
$(\lambda+3)^2=0$ y así tiene un único autovalor $\lambda=-3$ con un autovector
$v=
\left[\begin{matrix}
3\\
 1
\end{matrix}\right]$ 
Este sistema tiene dos soluciones una es de la forma $X_1=v_1e^{\lambda t}$
y la segunda es de la forma $X_2=v_1te^{\lambda t}+Pe^{\lambda t}$ donde $P$ es un vector. Para encontrar $P$ sustituimos en la ecuación $X'=AX$ y obtenemos 

$$(Av_1-\lambda v_1)te^{\lambda t}+(AP-\lambda P-v_1)e^{\lambda t}=0$$
puesto que $e^{\lambda t}$ es diferente de cero y el primer termino de la anterior es $(A-\lambda I)v_1=0$, entonce se reduce a 
$$(A-\lambda I)P=v_1$$


de esta forma una solución del sistema 
es $$X_1=\left[\begin{matrix}
3\\
 1
\end{matrix}\right]e^{-3t}$$
y la otra solución es 
$$\left[\begin{matrix}
3\\
 1
\end{matrix}\right]te^{-3t}+
\left[\begin{matrix}
p_1\\
 p_2
\end{matrix}\right]e^{-3t}
$$
para encontrar $\left[\begin{matrix}
p_1\\
 p_2
\end{matrix}\right]$ debemos resolver el sistema 

$$\left[\begin{matrix}
6&-18\\2&-6
\end{matrix}\right]\left[\begin{matrix}
p_1\\p_2
\end{matrix}\right]=\left[\begin{matrix}
3\\1
\end{matrix}\right]$$

Así obtenemos que $p_1=\frac{1}{2}$ y $p_2=0$ de esta forma 

$$X_2=\left[\begin{matrix}
3\\
 1
\end{matrix}\right]te^{-3t}+
\left[\begin{matrix}
\frac{1}{2}\\
 0
\end{matrix}\right]e^{-3t}
$$


de esta forma la soculción general es 

$$\left[\begin{matrix}
x(t)\\y(t)
\end{matrix}\right]=C_1\left\{\left[\begin{matrix}
3\\
 1
\end{matrix}\right]e^{-3t}\right\}
+C_2\left\{ \left[\begin{matrix}
3\\
 1
\end{matrix}\right]te^{-3t}+
\left[\begin{matrix}
\frac{1}{2}\\
 0
\end{matrix}\right]e^{-3t} \right\}
$$
