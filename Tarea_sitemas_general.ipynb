## Ejercicio 1

$\textbf{x}_1=\begin{bmatrix}e^{3t}\\0\\e^{3t}\end{bmatrix}$,

 $$\begin{split}\textbf{x}'&=\textbf{Ax},\\
 \begin{bmatrix}x_1'\\x_2'\\x_3'\end{bmatrix}&=\begin{bmatrix}1&-2&2\\ -2&1&2\\ 2&2&1\end{bmatrix}\begin{bmatrix}x_1\\x_2\\x_3\end{bmatrix}\end{split},$$ en $ 	\mathbb{R}$
 y que 
 $$\textbf{x}_p=\begin{bmatrix}5t+1\\2t\\4t+2\end{bmatrix},$$

 $$\begin{split}\textbf{x}'&=\textbf{Ax}+\textbf{f}(t),\\
